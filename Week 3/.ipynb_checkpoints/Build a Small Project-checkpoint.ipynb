# Build a Small Project Using Data Science Libraries

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Admin\Rasa_Intern\Files\WA_Marketing-Campaign.csv")
df.head()
```

```python
# Data Exploration

print("\nBASIC DATA EXPLORATION")
print("-" * 40)

print("\nDataset Shape:")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

print("\n\nFirst 5 rows:")
print(df.head())

print("\n\nDataset Info:")
print(df.info())

print("\n\nColumn Names:")
print(df.columns.tolist())

print("\n\nData Types:")
print(df.dtypes)

print("\n\nBasic Statistics:")
print(df.describe())

print("\n\nUnique Values in Categorical Columns:")
print(f"Market Sizes: {df['MarketSize'].unique()}")
print(f"Promotion Types: {df['Promotion'].unique()}")
print(f"Number of Markets: {df['MarketID'].nunique()}")
print(f"Number of Locations: {df['LocationID'].nunique()}")
```

```python
# Data Cleaning

print("\nDATA CLEANING")
print("-" * 40)

print("\nChecking for Missing Values:")
missing_values = df.isnull().sum()
print(missing_values)

if missing_values.sum() == 0:
    print("No missing values found!")
else:
    print(f"Total missing values: {missing_values.sum()}")

print("\nChecking for Duplicates:")
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

if duplicates == 0:
    print("No duplicate rows found!")

print("\nData Quality Check:")
print(f"Sales range: ${df['SalesInThousands'].min():.2f}K - ${df['SalesInThousands'].max():.2f}K")
print(f"Store age range: {df['AgeOfStore'].min()} - {df['AgeOfStore'].max()} years")
print(f"Week range: {df['week'].min()} - {df['week'].max()}")

# Check for any negative sales (data quality issue)
negative_sales = df[df['SalesInThousands'] < 0]
print(f"Negative sales entries: {len(negative_sales)}")

print("Data is clean and ready for analysis!")
```

```python
# BASIC ANALYSIS (GROUP BY, SORTING, FILTERING, AGGREGATIONS)

print("\nBASIC ANALYSIS")
print("-" * 40)

print("\nSales Performance by Market Size:")
market_analysis = df.groupby('MarketSize')['SalesInThousands'].agg([
    'mean', 'median', 'std', 'min', 'max', 'count'
]).round(2)
print(market_analysis)

print("\nSales Performance by Promotion Type:")
promotion_analysis = df.groupby('Promotion')['SalesInThousands'].agg([
    'mean', 'median', 'std', 'count'
]).round(2)
print(promotion_analysis)

print("\nWeekly Sales Trends:")
weekly_analysis = df.groupby('week')['SalesInThousands'].agg([
    'mean', 'sum', 'count'
]).round(2)
print(weekly_analysis)

print("\nTop 10 Performing Locations:")
top_locations = df.groupby('LocationID')['SalesInThousands'].mean().sort_values(ascending=False).head(10)
print(top_locations.round(2))

print("\nStore Age Impact on Sales:")
age_bins = pd.cut(df['AgeOfStore'], bins=4, labels=['New (1-6 yrs)', 'Young (7-12 yrs)', 'Mature (13-19 yrs)', 'Old (20+ yrs)'])
age_analysis = df.groupby(age_bins)['SalesInThousands'].agg(['mean', 'count']).round(2)
print(age_analysis)
```

```python
# DATA VISUALIZATIONS (MATPLOTLIB & SEABORN)

# Visualization 1: Bar Chart - Sales by Market Size
market_sales = df.groupby('MarketSize')['SalesInThousands'].mean()
plt.bar(market_sales.index, market_sales.values, color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
plt.title('Average Sales by Market Size', fontsize=14, fontweight='bold')
plt.xlabel('Market Size')
plt.ylabel('Sales (in Thousands)')
plt.xticks(rotation=45)
for i, v in enumerate(market_sales.values):
    plt.text(i, v + 1, f'${v:.1f}K', ha='center', fontweight='bold')
```

```python
# Visualization 2: Line Chart - Weekly Sales Trends

weekly_sales = df.groupby('week')['SalesInThousands'].mean()
plt.plot(weekly_sales.index, weekly_sales.values, marker='o', linewidth=3, markersize=8, color='#FF6B6B')
plt.title('Weekly Sales Trends', fontsize=14, fontweight='bold')
plt.xlabel('Week')
plt.ylabel('Average Sales (in Thousands)')
plt.grid(True, alpha=0.3)
```

```python
# Visualization 3: Box Plot - Sales Distribution by Promotion

sns.boxplot(data=df, x='Promotion', y='SalesInThousands', palette='Set2')
plt.title('Sales Distribution by Promotion Type', fontsize=14, fontweight='bold')
plt.xlabel('Promotion Type')
plt.ylabel('Sales (in Thousands)')
```

```python
# Visualization 4: Histogram - Sales Distribution

plt.hist(df['SalesInThousands'], bins=30, color='#4ECDC4', alpha=0.7, edgecolor='black')
plt.title('Distribution of Sales', fontsize=14, fontweight='bold')
plt.xlabel('Sales (in Thousands)')
plt.ylabel('Frequency')
plt.axvline(df['SalesInThousands'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: ${df["SalesInThousands"].mean():.1f}K')
plt.legend()
plt.show()
```

```python
# Visualization 5: Scatter Plot - Store Age vs Sales

colors = {'Small': '#FF6B6B', 'Medium': '#4ECDC4', 'Large': '#45B7D1'}
for market_size in df['MarketSize'].unique():
    subset = df[df['MarketSize'] == market_size]
    plt.scatter(subset['AgeOfStore'], subset['SalesInThousands'], 
               alpha=0.6, label=market_size, color=colors[market_size], s=50)
plt.title('Store Age vs Sales Performance', fontsize=14, fontweight='bold')
plt.xlabel('Age of Store (years)')
plt.ylabel('Sales (in Thousands)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Visualization 6: Pie Chart - Market Size Distribution

market_counts = df['MarketSize'].value_counts()
plt.pie(market_counts.values, labels=market_counts.index, autopct='%1.1f%%', 
        colors=['#FF6B6B', '#4ECDC4', '#45B7D1'], startangle=90)
plt.title('Distribution of Market Sizes', fontsize=14, fontweight='bold')
plt.show()
```

```python
# Visualization 7: Heatmap - Correlation Matrix

numeric_cols = ['MarketID', 'LocationID', 'AgeOfStore', 'Promotion', 'week', 'SalesInThousands']
correlation_matrix = df[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.2f', cbar_kws={'shrink': 0.8})
plt.title('Correlation Matrix', fontsize=14, fontweight='bold')
plt.show()
```

```python
# Visualization 8: Bar Chart - Top 10 Markets by Average Sales

top_markets = df.groupby('MarketID')['SalesInThousands'].mean().sort_values(ascending=False).head(10)
plt.bar(range(len(top_markets)), top_markets.values, color='#45B7D1')
plt.title('Top 10 Markets by Average Sales', fontsize=14, fontweight='bold')
plt.xlabel('Market Rank')
plt.ylabel('Average Sales (in Thousands)')
plt.xticks(range(len(top_markets)), [f'Market {i}' for i in top_markets.index], rotation=45)
plt.show()
```

```python
# Visualization 9: Violin Plot - Sales by Market Size

sns.violinplot(data=df, x='MarketSize', y='SalesInThousands', palette='Set1')
plt.title('Sales Distribution by Market Size', fontsize=14, fontweight='bold')
plt.xlabel('Market Size')
plt.ylabel('Sales (in Thousands)')
plt.show()
```

```python
#  INSIGHTS AND SUMMARY OF FINDINGS

print("\nKEY INSIGHTS AND FINDINGS")
print("=" * 60)

print("\nDATASET OVERVIEW:")
print(f" Total records analyzed: {len(df):,}")
print(f" Markets covered: {df['MarketID'].nunique()}")
print(f" Store locations: {df['LocationID'].nunique()}")
print(f" Time period: {df['week'].nunique()} weeks")

print("\nSALES PERFORMANCE INSIGHTS:")
avg_sales_by_market = df.groupby('MarketSize')['SalesInThousands'].mean()
print(f" Large markets perform best: ${avg_sales_by_market['Large']:.1f}K average sales")
print(f" Medium markets: ${avg_sales_by_market['Medium']:.1f}K average sales")
print(f" Small markets: ${avg_sales_by_market['Small']:.1f}K average sales")
print(f" Overall average sales: ${df['SalesInThousands'].mean():.1f}K")

print("\nPROMOTION EFFECTIVENESS:")
promo_effectiveness = df.groupby('Promotion')['SalesInThousands'].mean().sort_values(ascending=False)
best_promo = promo_effectiveness.index[0]
worst_promo = promo_effectiveness.index[-1]
print(f" Most effective promotion: Type {best_promo} (${promo_effectiveness[best_promo]:.1f}K avg)")
print(f" Least effective promotion: Type {worst_promo} (${promo_effectiveness[worst_promo]:.1f}K avg)")
print(f" Promotion impact: {((promo_effectiveness[best_promo] - promo_effectiveness[worst_promo]) / promo_effectiveness[worst_promo] * 100):.1f}% difference")

print("\nTEMPORAL TRENDS:")
weekly_trend = df.groupby('week')['SalesInThousands'].mean()
best_week = weekly_trend.idxmax()
worst_week = weekly_trend.idxmin()
print(f" Best performing week: Week {best_week} (${weekly_trend[best_week]:.1f}K)")
print(f" Lowest performing week: Week {worst_week} (${weekly_trend[worst_week]:.1f}K)")
print(f" Sales volatility: {df['SalesInThousands'].std():.2f}K standard deviation")

print("\nSTORE AGE ANALYSIS:")
age_correlation = df['AgeOfStore'].corr(df['SalesInThousands'])
print(f" Age-Sales correlation: {age_correlation:.3f}")
if abs(age_correlation) < 0.1:
    print(" Store age has minimal impact on sales performance")
elif age_correlation > 0:
    print(" Older stores tend to have higher sales")
else:
    print(" Newer stores tend to have higher sales")
```
