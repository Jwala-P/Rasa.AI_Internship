```python
import pandas as pd
```

```python
# Reading a CSV
time_series = pd.read_csv(r'C:\Users\Admin\Desktop\Rasa Intern\time_series.csv')
time_series.head()
```

```python
time_series.dtypes
```

```python
# Convert the Datetime column to pandas datetime object

pd.to_datetime(time_series.Datetime,dayfirst=True)
```

```python
# Create the following features using the Datetime feature (day of month, month, day of year, hour)

time_series['day of month']=pd.to_datetime(time_series.Datetime,dayfirst=True).dt.day
time_series['month']=pd.to_datetime(time_series.Datetime,dayfirst=True).dt.month
time_series['day of year']=pd.to_datetime(time_series.Datetime,dayfirst=True).dt.dayofyear
time_series['hour']=pd.to_datetime(time_series.Datetime,dayfirst=True).dt.hour

time_series.head()
```

```python
# DataFrame

sample_df = pd.DataFrame({
    'Id'   :  [1, 2, 3, 4, 5],
    'Name' :  [ 'Akash', 'Akansha', 'Naman', 'Pulkit', 'Anand'],
    'Marks':  [ 22, 23, 43, 41, 12]
})
sample_df
```

```python
# Sort the dataframe by Marks column in descending order

sample_df.sort_values('Marks',ascending=False)
```

```python
# DataFrame

city = pd.DataFrame({
    'Name' : ['Akash', 'Akansha', 'Pulkit', 'Anand', 'Naman'],
    'City' : ['Delhi', 'Gurugram', 'Gurugram', 'Delhi', 'Noida']
})
city
```

```python
# Find Marks of Student who live in Gurugram

Merged_df = pd.merge(sample_df,city, on ='Name')

Gurugram_students = Merged_df[Merged_df['City']=='Gurugram']
Gurugram_students[['Name','Marks','City']]
```

```python
# Big mart data

big_mart = pd.read_csv(r"C:\Users\Admin\Desktop\Rasa Intern\big_mart_sales.csv")
big_mart.head()
```

```python
# The number of unique Item_Type in each of the Outlet_Identifier

big_mart.groupby('Outlet_Identifier')['Item_Type'].nunique()
```

```python
# The average Item_Outlet_Sales for each category of Outlet_Size

big_mart.groupby('Outlet_Size')['Item_Outlet_Sales'].agg(['mean'])
```
