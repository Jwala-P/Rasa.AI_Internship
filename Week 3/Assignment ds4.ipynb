# Assignment Data Visualisation

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
# Reading a file

big_mart = pd.read_csv(r"C:\Users\Admin\Desktop\Rasa Intern\big_mart_sales.csv")
big_mart.head()
```

```python
# Sales By Outlet Size
Sales_by_OutletSize = big_mart.groupby('Outlet_Size').Item_Outlet_Sales.mean()
Sales_by_OutletSize

Sales_by_OutletSize.sort_values(inplace=True)

x = Sales_by_OutletSize.index.tolist()
y = Sales_by_OutletSize.values.tolist()

# Set xticks

positions = np.arange(len(x))
plt.xticks(ticks = positions,labels=x)
plt.bar(positions,y, color = ['red','orange','green'])
plt.show()
```

```python
# xlabels, ylabels, title

positions = np.arange(len(x))
plt.xticks(ticks = positions,labels=x)
plt.bar(positions,y, color = ['red','orange','green'])
plt.title('Mean Sales for each outlet type')
plt.xlabel('Outlet Size')
plt.ylabel('Sales')
plt.show()
```

```python
# Create a scatter plot to show the relative distribution of `Item_Weight` and `Item_Visibility`.

plt.figure(figsize=(20,8))
plt.scatter(data=big_mart,x='Item_Weight',y='Item_Visibility',color='red')
plt.xlabel('Item Weight')
plt.ylabel('Item Visibility')
plt.show()
```

```python
# Create a box plot of the variable Item_Outlet_Sales using the seaborn.

plt.figure(figsize=(5,8))
sns.boxplot(y='Item_Outlet_Sales',data=big_mart)
plt.title('Distribution of Item Outlet Sales')
plt.show()
```
