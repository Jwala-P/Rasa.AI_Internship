```python
# Final Task: Summary Report
# Load All CSV and JSON files

bm = pd.read_csv(r"C:\Users\Admin\Rasa_Intern\Files\big_mart_sales.csv")
bm.head()
```

```python
# Load a JSON file

json_data = pd.read_json(r'C:\Users\Admin\Rasa_Intern\Files\simple.json',lines=True)
json_data.head(2)
```

```python
# Perform at least 3 subsetting operations

# Subsetting By Value

bm1 = bm.loc[(bm['Outlet_Location_Type']=='Tier 1') & (bm['Item_Weight']>15)]
print("Dimensions of bm1:\n",bm1.shape, "\nDimensions of bm:\n", bm.shape)
bm1.head()
```

```python
# Subsetting By Label

bm2 = bm.loc[bm['Item_Identifier']=='FDN15']
bm2.head()
```

```python
# Subsetting by Index

bm3 = bm.iloc[4:7,[2,3,10]]
bm3
```

```python
# Creating a column called Units purchased to display the quantity of units purchased
bm['Units_Purchased'] = round((bm['Item_Outlet_Sales']/bm['Item_MRP']),0)
bm.head(2)
```

```python
# Creating a column to calculate the cost per kilo of each item

bm['Cost_per_kg'] = (bm['Item_MRP']/bm['Item_Weight'])
bm.head(2)
```

```python
# Renaming the Item_Identifier to Item_Code

bm.rename(columns={'Item_Identifier':'Item_Code'},inplace=True)
bm.head(2)
```

```python
# Droping NA values

bm.dropna(inplace=True)
bm.head()
```

```python
# Extract Insights

# Summary stats for each Fat Content category

# Changing Low Fat, low fat and LF into a standardised name and also Regular and reg into a standardised one as well
bm.loc[bm['Item_Fat_Content']=='low fat','Item_Fat_Content'] = "Low Fat"
bm.loc[bm['Item_Fat_Content']=='LF','Item_Fat_Content'] = "Low Fat"

bm.loc[bm['Item_Fat_Content']=='reg','Item_Fat_Content'] = "Regular"

Fat_Content_Stats = bm.groupby('Item_Fat_Content')['Item_MRP'].agg(['mean','max','min','median','std'])
Fat_Content_Stats.sort_values(by='mean',ascending=False)
```

```python
# Finding out which Item generated maximum revenue cummulatively

maximum_sale_price=bm.groupby('Item_Code')['Item_Outlet_Sales'].agg(['sum']).sort_values(by='sum',ascending=False)
maximum_sale_price.head(1)
```

```python
# Finding out the stats of the revenue generated based on each Item Type

Item_Type_Stats = bm.groupby('Item_Type')['Item_Outlet_Sales'].agg(['mean','min','max','median','std','count'])
Item_Type_Stats.sort_values(by='count',ascending = False)
```

```python
# Visualise a Trend

plt.figure(figsize=(22,8))
plt.xticks(fontsize=8)
plt.xlabel('Item Type',fontsize=18)
plt.ylabel('Mean Item Outlet Sales',fontsize=18)
sns.barplot(x='Item_Type',y='count',data=Item_Type_Stats,color='red')
plt.show()
```

```python
# Share of Sales per Outlet

# Group by Outlet
pie_data = big_mart.groupby('Outlet_Identifier')['Item_Outlet_Sales'].sum()

# Plot as pie chart
plt.figure(figsize=(8, 8))
plt.pie(pie_data, labels=pie_data.index,explode=[0,0,0.1,0,0],autopct='%1.1f%%', startangle=180, shadow=True,colors=plt.cm.Paired(np.linspace(0, 1, 5)))
plt.title("Sales Share by Outlet", fontsize=16)
plt.axis('equal')
plt.show()
```
